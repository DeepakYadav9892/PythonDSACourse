Lambda Functions in Python 

Lambda functions are small anonymous functions defined using the lambda keyword. 
They can have any number of arguments but only one expression . they are commenly used for short operations or as arguments to higher order functions .

```python
#syntax
lambda arguments:expression
```

```python
def addition(a,b):
    return a+b
```

```python
def addition(x,y,z):
    return x+y+z
addition(24,45,50)
```

```python
addition1=lambda x,y,z:x+y+z
addition1(12,13,24)
```
