```python
##variable length arguments 
##Positionnal and keywords arguments 

def print_numbers(*args):
    for  numbers in args:
        print(numbers)
```

```python
print_numbers(1,3,4,5,5,5,6,78,6,"Deepak")
```

```python
##Keywords arguments 
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

        
```

```python
print_details(name="krish",age="32",country="India")
```

```python
def  print_details(*args,**kwargs):
    for val in args:
        print(f"Positional arguments:{val}")
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,name="deepak",age="34",country="India")
```
