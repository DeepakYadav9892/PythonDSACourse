```python
##for loop 
for i in range(1,10,2): #intial , finnal , step
    print(i)
```

```python
##strings 
str="Deepak"
for i in str:
    print(i)
```

```python
##while loop 
##The while loop continues to execute  as long as the conditon is True.
count=0
while count<5:
    print(count)
    count=count+1
```

```python
count=0
while count%2==0:
    print(count)
    count=count+1
```

```python
##Loops control statements 
##Break statements exits the loop prematuraly
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
##continnue
##the continue stateemet skips the current itereationn and continues with nest .
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
#pass 
##the pas statement is null operation , it does nothing 
for i in range (5):
    if i==3:
        print("the number is ",i)
        pass
    print(i)
```

```python
##nested loops 
##a loop innside a loop
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
        
```

```python
#Examples -calculate the sum of first N natural numvbers usign a while and for loop 

##while loop 
n=10 
sum=0
count=1
while count<=n:
    sum=sum+count
    count=count+1
print("sum of first 10 natural number",sum)
```

```python
##for loop 
n=10 
sum=0
count=1

for i in range(11):
    sum=sum+i
print(sum)
```

```python
##Example-Prime number between 1 and 100
for num in range(1,101):
    if num>1:
        for i in range (2,num):
            if num%i==0:
                break
        else:
            print(num)
```
