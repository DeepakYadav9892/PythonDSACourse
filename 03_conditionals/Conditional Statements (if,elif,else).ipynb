```python
##else the elee statement execute teh block of code if the coditioninn the if statements is false 
age=17
if age>=18:
    print("you are eligible to vote ")
else:
    print("You are not eligible to vote ")
```

```python
##elif statement allows to cheack multiple conditions. it stands for else-if
age=17
if age<13:
    print("You are a child ")
elif age<18:
    print("You are a teeneager ")
else:
    print("you are adult")
```

```python
##Nested conditional Statments 
#You can place one or more if ,elif,or else statements inside another if,elif or statement  to create teh nested conndition statement 
##number even,odd , negative 

num=int(input("Enter the number:"))
if num>=0:
    print("the number  is positive")
    if num %2==0:
        print("The number is even ")
    else:
        print("the number is odd ")

else:
    print("the number is zero or negative ")
```

```python
##usecases , practical exmination 

##if you need determine the given year is leap year using nested condtion statement

year=int(input("Enter the year"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"is a leap year")
        else:
            print(year,"is not a leap year")
    else:
        print(year,"is a leap year ")
else:
    print(year,"is not a leap year ")
```
