2)Analysis by Asymptotic Notation  
Big-O Notation : Describes the upper bound or the worst-case time/space complexity of an algorithm. It guarantees that the algorithm's growth rate will be no worse than a certain function (e.g., \(O(n)\), \(O(n^{2})\)).

Big-Omega Notation :Describes the lower bound or the best-case complexity of an algorithm. It guarantees that the algorithm's growth rate will be no better than a certain function.
Big-Theta Notation : Describes the tight bound (both the upper and lower bound) of an algorithm. It indicates that the algorithm's performance is precisely characterized by a specific growth function </span>
