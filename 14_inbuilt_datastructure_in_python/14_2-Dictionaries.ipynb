Introduction to dictionaries 

dictionaries are unorderd collections of items. They sotre data inn key 
values pairs . keys must be unique and immutable (eg strings, numbers, or tuples ),
while values can be of any type


```python
##creating Dictionaries 
empty_dict={}
print(type(empty_dict))


```

```python
empty_dict=dict()
print(empty_dict)
```

```python
student={"name":"Deepak","Age":25,"grade":24}
print(student) ##key value pairs 
```

```python
#error 
#single key is always used in dictionaris 

student={"name":"Deepak","Age":25,"name":24}
print(student) ##key value pairs 
```

```python
#accessing dictionary elements 
student={"name":"Deepak","Age":25,"grade":24}
print(student) ##key value pairs 
student["grade"]
print(student["Age"])

#accesing value usign get() method 
print(student.get("grade"))
print(student.get('name'))
print(student.get('name','grade'))
```

```python
##Modifying dictionary elements 
##dictionary are mutable in nature you can add, update , or delete elements 
print(student)
```

```python
student["Age"]=33
print(student)
student["adress"]="Mumbai"
print(student)
```

```python
del student['grade']
print(student)
```

```python
##common dictionary methods 
keys=student.keys()
values=student.values()
print(keys)
print(values)

items=student.items()## get all key values pairs  in the form of tuple 

print(items)
```

```python
#shallow copy
student_copy=student
print(student)
print(student_copy)
student["name"]="Deepak1"
```

```python
student["name"]="Deepak2"
print(student)
print(student_copy)

student_copy1=student.copy() #shallow copy 
print(student_copy1)
```

```python
##iterating over dictionaries 
##you can cuse loops to iterate over dictionaries, keys , values or items 

##Iterating overkeys 
for keys in student.keys():
    print(keys)
```

```python
#Iterate over values 
for values in student.values():
    print(values)
```

```python
#iterate over key value pairs 
for key, value in student.items():
    print(f"{key}:{value}")
```

```python
##nested dictionaries : dictionaries inside anotehre disctionaryy 
students={
    "student1":{"name":"krish","age":32},
    "student2":{"name":"petter","age":35}
}

print(students)
```

```python
#access nested dictionaries elements 
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
##Iterating over nested dictionaries 
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        
        print(f"{key}:{value}")
```

```python
#Dictionaries comprehesionns 
squares={x:x**2 for x in range(10)}
print(squares)
```

```python
#Condition dictionary comprehension 
events={x:x**2 for x in range(10) if x%2==0}
print(events)
```

```python
##Practical example 
##use a dictionary to count the frequency of elements inlist 
numbers=[1,3,4,3,4,5,5,6,77,7,7,8]
frequency={
}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
##Merge 2 dictionaries into one 
dict1={"a":1,"b":2}
dict2={"b":3,"c":4}

merged_dict={**dict1,**dict2}
print(merged_dict)
```
