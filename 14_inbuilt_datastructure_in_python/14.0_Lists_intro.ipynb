##list comprehesions

##Basics syntax [Expression for item in iterable]

with conditionnal logic [expression for item in iterable if condition ]


Nested List comprehensions 
[ expression for item in iterable1 for item2 in iterable2]

##

```python
##basic list comprehension 
sqaure=[num**2 for num in range(10)]
print(sqaure)
```

```python
##List comprehension with conditionn 
lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)
print(lst)
```

```python
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
##Nested list comprehesion

lst1=[1,2,3,4,5]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]

print(pair)
```

```python
##List comprehenstionn with function calls 
words=["hello","world","Python","list","comprehensionn"]
length=[len(word)for word in words]
print(length)
```
