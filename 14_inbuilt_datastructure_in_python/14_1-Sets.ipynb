Sets :

Sets are a built in data types in python used to store collections of nique items . they are unorderd , meaning that the elements do no follow a specific oreder, and tehy do no allow duplicate elements . Sets are useful for membership tests, eleminating duplcate entries, and performin mathematical set operationn like uion intersection, difference and symmetic diffrence . 


```python
##create a set 
my_set={1,2,3,4,5}
print(my_set)
my_empty_set=set()
print(type(my_empty_set))
```

```python
my_set={1,2,3,4,5,6,6}
print(my_set)
```

```python
##Besics set operation 
##Adding and removing elements 
my_set.add(45)
my_set
#remove elements
my_set.remove(45)
print(my_set)
my_set.remove(5)
```

```python
my_set.discard(5)
print(my_set)
##pop method
remove_element=my_set.pop()
print(remove_element)
print(my_set)
```

```python
###Clear all the elements 
my_set.clear()
print(my_set)
```

```python
##Set Memberhsip test 
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
#Methametical operation 
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}
##Union 
union_set=set1.union(set2)
print(union_set)


##Intesecttion 
instersection_set=set1.intersection(set2)
print(instersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1={1,3,4,5,6,7}
set2={4,5,6,7,8,9}

##Difference 
print(set1.difference(set2))
```

```python
set2.difference(set1)
```

```python
##symmetric diffrence  commmon element we gget from it 

set1.symmetric_difference(set2)

```

```python
##sets method 
set1={1,2,3,4,5} 
set2={3,4,5,6}

print(set1.issubset(set2))
print(set1.issuperset(set2))
```
