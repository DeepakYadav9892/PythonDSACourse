Example1. Manage A to DO llist 

create to do list to keep track of tasks

Example2 Organising Students Grades 
. Create list to store and calculate avereage grades for students 

```python
#Organising student grades 
grades=[85,92,78,90,88]
#adding a new grade 
grades.append(95)
#calculating the averarge grade 

average_grade=sum(grades)/len(grades)
print(f"Average Grade:{average_grade:.2f}")
#Finding the highest and lowest grades 
highest_grade=max(grades)
lowest_grade=min(grades)
print(f"Highest grade:{highest_grade}")
print(f"Lowest grade:{lowest_grade}")
```
