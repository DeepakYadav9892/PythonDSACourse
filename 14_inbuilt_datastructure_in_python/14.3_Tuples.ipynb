```python
print(mixed_tuple)
```

```python
##Acesssing tuple elements 
numbers[0]
numbers[::-1]

##tuple operations
mixed_tuple
##Tuple Operations
concatenation_tuple=numbers+mixed_tuple
print(concatenation_tuple)
```

```python
##Imutable Nature of tuples 
##Tuples are aimmutable , meaning their elements canot be changed once assigned 
lst=[1,2,3,4,5]
print(lst)
lst[1]="Deepak"
print(lst)
```

```python
#Tuple maethods 
print(numbers.count(1))
print(numbers.index(3))
```

```python
#packing and unpacking tuple 
packed_tuple1=1,"hello",3.24
print(packed_tuple1)

##unpacking a tuple 
a,b,c=packed_tuple1
print(a)
print(b)
print(c)
```

```python
#unpacking with * 
numbers=(1,2,3,4,5,6,7)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
#nested tuplle 
lst=[[1,2,3,4],[5,6,7,8],[1,"hello",3.14,"c"]]
lst[0][0:3]
```

```python
nested_tuple=((1,2,3),("a","b","c"),(True,False))
##access the elements inside a tuple 

print(nested_tuple[0])
```

```python
#iterating over nested tuples 
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
```
